## Линейный перенос разностной схемой

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nx = 101
dx = 2/(nx-1)
nt = 25
dt = .02
c = 1

x = np.linspace(0, 2, nx)
u = np.ones(nx)

u[np.intersect1d(np.where(x >= 0.5), np.where(x <= 1))] = 2
plt.plot(x, u)
```

```python
for n in range(1, nt):
    un = u.copy()
    for i in range(1, nx):
        u[i] = un[i] - c*dt/dx*(un[i] - un[i-1])
```

```python
plt.plot(x, u)
```

```python
def linearconv(nx):
    dx = 2/(nx-1)
    nt = 25
    dt = .02
    c = 1

    x = np.linspace(0, 2, nx)
    u = np.ones(nx)

    u[np.intersect1d(np.where(x >= 0.5), np.where(x <= 1))] = 2
    
    for n in range(1, nt):
        un = u.copy()
        for i in range(1, nx):
            u[i] = un[i] - c*dt/dx*(un[i] - un[i-1])
            
linearconv(81)
plt.plot(x, u)
```
