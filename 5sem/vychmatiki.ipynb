```python
#Нарисовать решение задачи du/dx = x, u(0) = 0 
#при аппроксимации производной методом центральных разностей

import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(0., 1., 10000)
h = x[1]-x[0]
u = np.zeros_like(x)
unew = u.copy()

eps = 1.0e-4
j = 0

while(True):
    maxdelta = 0.
    u[0] = 0.
    
    for i in range(1, len(x)-1):
        unew[i+1] = x[i]*2*h + u[i-1]
    j+=1
    
    for i in range(len(x)):
        if (abs(unew[i] - u[i]) > maxdelta):
            maxdelta = abs(unew[i] - u[i])
            
    if j==3000:
        break
        
    u = unew.copy()

plt.plot(x, [xi**2/2 for xi in x], label='точное решение', color='k', linestyle='--')
plt.scatter(x, u, label='разностная схема', s=20)

plt.legend()
plt.grid()
plt.axis('equal')

print("Решение сошлось за", j, "шагов")
```
