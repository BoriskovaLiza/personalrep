```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
img = mpimg.imread('plate_heat_transfer_crop.png')[::-1,:]
plt.imshow(img, origin='lower')
plt.show();
```

```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Button

fig, ax = plt.subplots()
plt.imshow(img, origin='lower')

fig = plt.gcf()
ax = plt.gca()

fig.set_size_inches(9, 7)
text = ax.text(0,0, "", va="bottom", ha="left")

X = []
Y = []

plt.show()

plt.imshow(img, origin='lower')
fig = plt.gcf()
fig.set_size_inches(9, 7)
ax = plt.gca()
text = ax.text(0,0, "", va="bottom", ha="left")

def onclick(event):
    tx = 'button=%d, x=%d, y=%d, xdata=%.2f, ydata=%.2f' % (event.button, event.x, event.y, event.xdata, event.ydata)
    text.set_fontsize(16)
    text.set_text(tx)
    X.append(event.xdata/img.shape[1]*0.3)
    Y.append(event.ydata/img.shape[0]*25)

cid = fig.canvas.mpl_connect('button_press_event', onclick)
```

```python
xmin, xmax = 0.0, 0.3
ymin, ymax = 0.0, 25

fig, ax = plt.subplots()
ax.set(xlim=(xmin, xmax), ylim=(ymin, ymax))

plt.grid(True)
ax.plot(X, Y, color='k', lw=2)
fig.set_size_inches(9, 6)
plt.plot(X, Y, 'bo')
```

```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Button

fig, ax = plt.subplots()
plt.imshow(img, origin='lower')

fig = plt.gcf()
ax = plt.gca()

fig.set_size_inches(9, 7)

text = ax.text(0,0, "", va="bottom", ha="left")

X = [0]
Y = [0]
xzero = 0
yzero = 0

plt.show()

plt.imshow(img, origin='lower')
fig = plt.gcf()
fig.set_size_inches(9, 7)
ax = plt.gca()
text = ax.text(0,0, "", va="bottom", ha="left")

class Index(object):
    ind = 0

    def lb_corner(self, event):
        global X, Y
        X.pop()
        Y.pop()
        X[0] = X[-1]
        Y[0] = Y[-1]
        X.pop()
        Y.pop()

callback = Index()
axb = plt.axes([0.7, 0.05, 0.1, 0.075])
bzero = Button(axb, '0:0 pressed')
bzero.on_clicked(callback.lb_corner)

def onclick(event):
    tx = 'button=%d, x=%d, y=%d, xdata=%.2f, ydata=%.2f' % (event.button, event.x, event.y, event.xdata, event.ydata)
    text.set_fontsize(16)
    text.set_text(tx)
    X.append(event.xdata)
    Y.append(event.ydata)

cid = fig.canvas.mpl_connect('button_press_event', onclick)
```

```python
xmin, xmax = 0.0, 0.3
ymin, ymax = 0.0, 25

X[:] = X[:] - X[0]
Y[:] = Y[:] - Y[0]
X[:] = X[:]/X[-1]*0.3
Y[:] = Y[:]/Y[-1]*25

X.pop()
Y.pop()
del X[0]
del Y[0]

fig, ax = plt.subplots()
ax.set(xlim=(xmin, xmax), ylim=(ymin, ymax))

plt.grid(True)
ax.plot(X, Y, color='k', lw=2)
fig.set_size_inches(10, 6)
plt.plot(X, Y, 'bo')
```
